# Students Social Media Addiction Data Analysis

This notebook analyzes social media usage patterns among students and visualizes key insights using matplotlib.


```python
# Importing useful libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

## Importing and reading data from CSV file


```python
# Load the dataset
data = pd.read_csv("Students Social Media Addiction.csv", index_col="Student_ID")
data.head()
```

## Data Cleaning and Inspection


```python
# Check for missing values and data types
data.info()
data.describe()
```

## Exploratory Data Analysis (EDA)


```python
# Unique values in key columns
print("Unique Countries:", data["Country"].unique())
print("Unique Platforms:", data["Most_Used_Platform"].unique())
print("Unique Academic Levels:", data["Academic_Level"].unique())
```

```python
# Distribution of Average Daily Usage Hours
plt.figure(figsize=(7, 5))
plt.hist(data["Avg_Daily_Usage_Hours"], bins=20, color="skyblue", edgecolor="black")
plt.title("Distribution of Average Daily Social Media Usage (Hours)")
plt.xlabel("Avg Daily Usage Hours")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Top 5 Most Used Social Media Platforms (Pie Chart)
platform_counts = data["Most_Used_Platform"].value_counts().head()
plt.figure(figsize=(7, 7))
plt.pie(
    platform_counts,
    labels=platform_counts.index,
    autopct="%1.1f%%",
    startangle=140,
    explode=[0.05] * len(platform_counts),
    shadow=True,
    colors=plt.cm.Paired.colors,
)
plt.title("Top 5 Most Used Social Media Platforms")
plt.axis("equal")
plt.show()
```

```python
# Country-wise Student Count (Top 10 Countries Only)
country_counts = data["Country"].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.barh(country_counts.index[::-1], country_counts.values[::-1], color="teal")
plt.title("Top 10 Countries by Number of Students")
plt.xlabel("Number of Students")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
# Boxplot: Academic Level vs. Average Daily Usage Hours
plt.figure(figsize=(8, 6))
sns.boxplot(x="Academic_Level", y="Avg_Daily_Usage_Hours", data=data)
plt.title("Average Daily Social Media Usage by Academic Level")
plt.xlabel("Academic Level")
plt.ylabel("Avg Daily Usage Hours")
plt.show()
```

```python
# More descriptive statistics for all columns
display(data.describe(include="all"))
```

```python
# Gender distribution (if available)
if "Gender" in data.columns:
    gender_counts = data["Gender"].value_counts()
    plt.figure(figsize=(5, 4))
    sns.barplot(x=gender_counts.index, y=gender_counts.values, palette="pastel")
    plt.title("Gender Distribution")
    plt.ylabel("Number of Students")
    plt.xlabel("Gender")
    for i, v in enumerate(gender_counts.values):
        plt.text(i, v + 1, str(v), ha="center")
    plt.show()
```

```python
# Correlation heatmap for numeric columns
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
